```python
# Import libraries
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
```

# Demonstrating Projection Distortions with Cartopy

#### Sources: https://pro.arcgis.com/en/pro-app/latest/help/mapping/properties/list-of-supported-map-projections.htm, https://scitools.org.uk/cartopy/docs/v0.15/crs/projections.html#mollweide




### Plate Carr√©e is one of the oldest and most basic projections. It uses square grid of latitudes and longitudes. Distortion increases as you get further from the equator.

To plot a basic world map with a projection, we set up a simple matplotlib plot, setting 'projection=ccrs.' followed by our desired projection name. Ccrs stands for Cartopy Coordinate Reference System. Options available in Cartopy can be found here: https://scitools.org.uk/cartopy/docs/v0.15/crs/projections.html. To add geographic features such as coastlines, country borders, and oceans using Cartopy, we use cfeature followed by the feature type. 

```python
# Create a figure of the specified size
fig = plt.figure(figsize=(11, 8.5)) 

# Specify that we want 1 subplot with the Plate Carree projection
ax = plt.subplot(1,1,1, projection=ccrs.PlateCarree())

# Set the figure title
ax.set_title('Plate Carr√©e', size=18)

# Add coastlines, country borders, and oceans to our plot using Cartopy
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)

# Display the plot
plt.show()
```

### The Mercator projection is a cylindrical projection. Area is extremely distorted at high latitudes. For example, Greenland appears to be the same size as Africa, which we know isn't true.

```python
# Create the same figure, but using the Mercator projection
fig = plt.figure(figsize=(11, 8.5))
ax = plt.subplot(1,1,1, projection=ccrs.Mercator())
ax.set_title('Mercator', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
plt.show()
```

### Molleweide is an equal area projection. Distortion is particularly large near the projection edges. 

```python
fig = plt.figure(figsize=(11, 8.5))
ax = plt.subplot(1,1,1, projection=ccrs.Mollweide())
ax.set_title('Molleweide', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
plt.show()
```

### The Sinusoidal projection is an equal area projection. Distortion near the center of the world map is relatively  mild, but increases greatly towards the projection edges. 

```python
fig = plt.figure(figsize=(11, 8.5))
ax = plt.subplot(1,1,1, projection=ccrs.Sinusoidal())
ax.set_title('Sinusoidal', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
plt.show()
```

### The Gnomonic projection has uses for navigation, but it clearly distorts area, shapes, and distance everywhere except right near the center. 

```python
fig = plt.figure(figsize=(11, 8.5))
ax = plt.subplot(1,1,1, projection=ccrs.Gnomonic())
ax.set_title('Gnomonic', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
plt.show()
```

### Conic projections are generally preferred for the CONUS (and many other mid-latitude regions) to minimize distortion. These include Albers Equal Area Conic and Lambert Conformal Conic, both of which are shown below.

```python
fig = plt.figure(figsize=(11, 8.5))

# Use the Albers equal area projection and specify the central longitude as -96 degrees to center the map on the CONUS
ax = plt.subplot(1,1,1, projection=ccrs.AlbersEqualArea(central_longitude=-96))
# Set the figure title; add coastlines, country borders, oceans, and states to the map using Cartopy
ax.set_title('Albers Equal Area', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.STATES)
# Specify the extent of our figure using latitude/longitude; 
# crs=crs=ccrs.PlateCarree() specifies that we're giving these coordinates in latitude/longitude
ax.set_extent([-120, -73, 22.5, 52.5], crs=ccrs.PlateCarree())
plt.show()
```

```python
# Same plot as above, but using Lambert Conformal projection this time
fig = plt.figure(figsize=(11, 8.5))
ax = plt.subplot(1,1,1, projection=ccrs.LambertConformal(central_longitude=-96))
ax.set_title('Lambert Conformal', size=18)
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.STATES)
ax.set_extent([-120, -73, 22.5, 52.5], crs=ccrs.PlateCarree())
plt.show()
```
